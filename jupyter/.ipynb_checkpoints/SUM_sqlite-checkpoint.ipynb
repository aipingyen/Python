```python
import sqlite3
import pandas as pd
conn = sqlite3.connect('./SUM.db')
cursor = conn.cursor()
cursor.execute("SELECT brand, color, cc, mileage, years, price FROM SUM2 where price != -1 and brand = 'TOYOTA' ")
for row in cursor:
    print(row)
cursor.close()
```

```python
from sqlalchemy import create_engine
df = pd.read_sql("SELECT brand, color, cc, mileage, years, price FROM SUM2 where price != -1 and brand = 'TOYOTA' ;", create_engine('sqlite:///SUM.db', echo=True))
df
```

```python
conn = sqlite3.connect('./SUM.db')
cursor = conn.cursor()
cursor.execute("SELECT brand, count(*) FROM SUM2 where price != -1 group by brand")
for row in cursor:
    print(row)
cursor.close()
```
