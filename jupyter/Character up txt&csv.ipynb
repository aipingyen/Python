```python
#去除中間空白Character
das = []
a=""
with open("C:\\Users\\BIG DATA\\Desktop\\dict2.txt","r") as fr:
    data=fr.read().split("\n")
#     print(data)
    for da in data:
        for d in da:
            a += d.strip("    ").strip("\t")
            
        das.append(a)
        a=""
        
print(das)

with open("C:\\Users\\BIG DATA\\Desktop\\dict3.txt","w") as fw:
    fw.write('\n'.join(das))
```
