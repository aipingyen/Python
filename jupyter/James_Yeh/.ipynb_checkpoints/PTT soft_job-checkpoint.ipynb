```python
import requests
from bs4 import BeautifulSoup
import re
from collections import Counter
```

```python
URL = "https://www.ptt.cc/"
```

```python
res = requests.get(URL + "bbs/Soft_Job/index.html" )
soup = BeautifulSoup(res.text , 'lxml')
```

```python
bottons = soup.select('a.btn.wide')
bottons
```

```python
totalpage = int(bottons[1]['href'].split('index')[1].split('.')[0])+1
totalpage
```

```python
#main-content > div:nth-child(4)
#main-content > div:nth-child(3) > span.article-meta-value
#main-container > div.r-list-container.action-bar-margin.bbs-screen > div:nth-child(14) > div.title > a
#main-content > div:nth-child(4) > span.article-meta-value
#main-content > div:nth-child(3)
#main-content
```

```python
with open('./lan.text') as f:
    lan_text = f.read().split('\n')
lan_text
```

```python
lc = Counter()
for page in range(totalpage,1,-1):
    try:    
        res = requests.get(URL+"bbs/Soft_Job/index{}.html".format(page))
        soup = BeautifulSoup(res.text,'lxml')
        links = soup.select('div.title > a')
        for link in links:
    #         print(link)
            title = link.text
    #         print(title)
            if("[徵才]" in title):
                article_URL = URL +link['href']
                article_res = requests.get(article_URL)
                article_soup = BeautifulSoup(article_res.text,'lxml')
    #             print(article_soup)
                article_dt = article_soup.select('span.article-meta-value')[3].text
                print(article_dt)            
                if((re.match('.*\s201[2-7]', article_dt)[0]) != None ):
    #                 print('1')
                    main_content = article_soup.select_one("#main-container > div")
    #                 print(main_content)
                    for div in main_content.find_all('div'):
                        div.decompose()
                    for a in main_content.find_all('a'):
                        a.decompose()
#                     for script in main_content.find_all('script'):
#                         script.decompose()
                    for span in main_content.find_all('span'):
                        span.decompose()
#                     print(main_content)
                    word = re.findall('[A-Za-z]+[+#?]*' , main_content)
                    for ind in word:
                        if ind in lan_text:
                            if ind in lc:
                                lc[ind] += 1
                            else:
                                lc[ind] = 1
                   
    except:
        continue
    print('[Debug] Page {} is finished'.format(totalpage))
lc.most_common()   
```
