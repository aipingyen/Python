# 2.若無flash字樣,則為JS檔,繼續檢查原始碼network/heading資訊區

```python
import re #使用正規表示法搜尋
m= re.search('"data":\[\[.*\]\]',res.text)   #正規表示法搜尋 data:[[ 括號中之資料 ]]
m.group()
```

```python
m= re.search('"data":\[\[.*\]\]',res.text)   
print('{'+m.group()+'}')       #調整為jason格式
```

```python
import json
data=json.loads('{'+m.group()+'}')
print(data)                      #轉存jason格式
#JSON = json.loads(json_format)
```
