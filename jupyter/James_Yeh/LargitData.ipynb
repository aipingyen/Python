<h2>list to dict

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(text,'lxml')
# print(soup.select('.td2bg1.arWebFont'))
com = soup.select('.td2bg1.arWebFont')   #list
com_list=[]
com_list2=[]
com_list3=[]
for comind in com :
    com_name = comind.text.strip()    #str
    com_list3.append(com_name)
#     print(com_name) 
for ind, row in enumerate(com_list3):
    if ind%2==1:
        com_list.append(row)
    if ind%2==0:
        com_list2.append(row)
    com_dict = {k:v for k,v in zip(com_list, com_list2)}
com_dict
```
