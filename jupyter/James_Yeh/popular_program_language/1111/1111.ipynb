# Save all links

# Word Count


```python
from collections import Counter
counter = Counter()
with open('1111_job.csv', 'r') as fr:
    for line in fr:
        if len(line) > 0:
            words = list(set(line.strip().split('/')))
            for word in words:
                if word.lower() in counter.keys():
                    counter[word.lower()] += 1
                else:
                    counter[word.lower()] = 1
```

```python
with open('1111.csv', 'w') as f:
    for lan,counts in counter.most_common():
        f.write('{},{}\n'.format(lan,counts))
```
