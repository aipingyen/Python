```python
from bs4 import BeautifulSoup
import requests as r
for i in range(1,71):
    Host = 'https://www.1111.com.tw'
    url = Host+'/job-bank/job-index.asp?si=1&ss=s&d0=140200,140300,140400&ps=100&page={}'.format(i)
    res = r.get(url)
    soup = BeautifulSoup(res.text, 'lxml')
    links = ['https:'+link['href'] for link in soup.select('div.jbInfo > div > h3 > a')]
    with open('link_1111.csv', 'a') as f:
        f.write('\n'.join(links))
    print('[Debug] Page {} is finished'.format(i))
```

# Thread

```python
import time
import threading
with open('link_1111.csv', 'r') as fr:
    link_list = fr.read().strip().split('\n')
with open('1111_job.csv', 'a') as f:
    for idx, link in enumerate(link_list):
        if idx % 10 == 0:
            print('Cuttent page {}'.format(idx))
        count = 5
        while count:
            try:
                lock = threading.Lock()
                th = threading.Thread(target=save_content, args=(link, lock, f))   # (target function , args)
                th.start()
                th.join()
                break
            except Exception as e:
                print(e)
                time.sleep(0.3)
                count -= 1
                continue 
            
```

# Word Count


```python
from collections import Counter
counter = Counter()
with open('1111_job.csv', 'r') as fr:
    for line in fr:
        if len(line) > 0:
            words = list(set(line.strip().split('/')))
            for word in words:
                if word.lower() in counter.keys():
                    counter[word.lower()] += 1
                else:
                    counter[word.lower()] = 1
```

```python
with open('1111.csv', 'w') as f:
    for lan,counts in counter.most_common():
        f.write('{},{}\n'.format(lan,counts))
```
