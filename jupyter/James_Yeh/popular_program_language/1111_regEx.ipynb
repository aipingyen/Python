```python
import requests as r
from bs4 import BeautifulSoup
import re
from collections import Counter
import pandas as pd
```

```python
#改用正規化表示
with open('./lan.text') as f:
    lan_text = f.read().upper().split('\n')
lan_text[:10]
```

```python
lc = Counter()
count = 0
for i in range(1,151):
    try:
        Host = 'https://www.1111.com.tw'
        url = Host+'/job-bank/job-index.asp?si=1&ss=s&d0=140200,140300,140400&ps=100&page={}'.format(i)
        res = r.get(url)
        soup = BeautifulSoup(res.text,'lxml')
        rows = soup.select('div.jbInfo > div > h3 > a')
        
        for row in rows:
            word_list = []
            url2 = 'https:'+row['href']
            print(url2)
            res2 = r.get(url2)
            soup2 = BeautifulSoup(res2.text,'lxml')
            
            content = soup2.select_one('dl > dd > p')
#             for idx in content:
            word = re.findall('object c|visual basic|\.net|[A-Za-z]+[+#]*' , content.text, re.IGNORECASE)
            word_list += word
        
            content2 = soup2.select_one('dl > dd')
            for idx in content2:
                for a in idx.find_all('a'):
                    a.decompose()
            for idx2 in content2:
                word2 = re.findall('object c|visual basic|\.net|[A-Za-z]+[+#]*' , idx2.text, re.IGNORECASE)
                word_list += word2
                
#             print(word_list)
            pure_word = list(set(word_list))
            print(pure_word)
            for ind in pure_word:
#                 if ind in lan_text:
                if ind in lc:
                    lc[ind] += 1
                else:
                    lc[ind] = 1
#             print(lc.most_common())
            count += 1
            if (count%20==0):
                print("COUNT = {}".format(count))
    except Exception as e:
        print(e)
        continue
    print(lc.most_common())
    
    print('[Debug] Page {} is finished'.format(i)) 
lc.most_common()
```

```python
with open('1111_2.csv', 'w') as f:
    for lang, counts in lc.most_common():
        f.write('{},{}\n'.format(lang,counts))
```
