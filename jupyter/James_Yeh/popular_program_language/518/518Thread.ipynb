```python
import requests
from bs4 import BeautifulSoup
import re
from collections import Counter
global static #增加一個全域變數用來紀錄統計結果
import threading
static={}
```

```python
import _thread
import time
```

```python
with open('./lan.text') as f: #打開我們要比對的字典檔
    wordlen = f.read().split('\n')
del wordlen[0]
wordlen[:10] 
```

```python
def innerURL(url): #一個用來分析內頁的function
    
    inner = requests.get(url)  #用get連到網頁
    innersoup = BeautifulSoup(inner.text, 'lxml')  #放到BeautifulSoup解析
    innerselect = str(innersoup.select('div.job-detail-box > dl'))  #.upper()  
    #找到可能會出現找尋目標的區塊將全部英文變大寫

    lan1=re.findall('[A-Z]+[+#?]*' , innerselect)
    #用正規法抓出每個英文單字
    
#     #將兩個list合併
#     lan1.extend(lan2)
#     #排除調內文出現的雜訊
#     if 'HTML#' in lan1:
#         lan1.remove('HTML#')
#     if 'HTML?' in lan1:
#         lan1.remove('HTML?')
   
    lan1 = list(set(lan1))     #list(set( list ))
    #清除調list內重複的部份
    
    for i in lan1: 
        if i in wordlen: 
            if i in static: 
                static[i] += 1
            else:
                static[i] = 1 
    return static
```

```python
def pageURL(url):  
    res = requests.get(url)
    soup = BeautifulSoup(res.text, 'lxml')
    links = soup.select('li.title > a')
    for link in links:
        innerURL(str(link['href']))  #throw in innerURL function
    return static
    
```

```python
class MyThread (threading.Thread): # 繼承 Thread 類別
    def __init__(self,number):
        threading.Thread.__init__(self)
        self.number=number
    def run(self): 
        pageURL('https://www.518.com.tw/job-index-P-'+str(self.number)+'.html?i=1&am=1&ab=2032001,2032002,') 
        #throw in pageURL function
        print('[Debug] Page {} is finished'.format(self.number))
```

```python
threads=[]
# wordCoun = Counter()
for i in range(1,55,1): 
    Thread=MyThread(i)
    threads.append(Thread)
for thread in threads:
    thread.start()
for thread in threads:
    thread.join()
```
