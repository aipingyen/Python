```python
import sqlite3
import sys, math
db = sqlite3.connect('./SUM.db')
cur = db.cursor()
# for row in c.execute('select * from HOT limit 1;'):
data = []
for row in cur.execute('select * from SUM;'):
    data.append(row)
# print(data)
db.commit()
db.close()

before = ['安全氣囊','倒車影像顯示','倒車雷達','車前距雷達','電動後視鏡','免KEY','HID','全景天窗','衛星導航','DVD','電動座椅','TCS','定速系統','鋁圈','ABS','皮椅','方向盤換檔控制','方向盤音響控制']
after  = ['安全氣囊','倒車影像','倒車雷達','ACC主動式定速系統','電動後視鏡','keyless免鑰系統','HID頭燈','天窗','衛星導航','DVD','電動座椅','TCS循跡系統','定速','鋁圈','ABS','皮椅','換檔撥片','多功能方向盤']
total = len(data)
data_dict = list(zip(before,after))
count = 0
# print(data[1][9])
for each_data in data:
    change = each_data[9]
    for each_equip in data_dict:    
        change = change.replace(each_equip[0],each_equip[1])
    printstring = ('now goes on' + str(count) + ', which is ' + str(math.floor(int(count+1)/int(total)*100)) + str('% finished'))
    sys.stdout.write('\r' + printstring )
    conn = sqlite3.connect('./SUM.db')
    c = conn.cursor()
    t = (change,each_data[1])
    c.execute('UPDATE SUM SET equip = ? WHERE url = ?;', t)
    conn.commit()
    conn.close()
    count+=1
print('\n data clean over!')
```
