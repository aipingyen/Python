```python
import requests as r
from  bs4 import BeautifulSoup
from lib.toolbox import gen_header
import math
import re
import random
import time
header_str = 'User-Agent:Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/57.0.2987.133 Safari/537.36'
HEADER = gen_header(header_str)
carName_list = ['PORSCHE','NISSAN','TOYOTA','MITSUBISHI','HONDA','FORD','AUDI','MERCEDES BENZ','BMW','LEXUS','VOLKSWAGEN','MAZDA','SUBARU','SUZUKI','VOLVO']
host = 'https://www.mobile01.com/'
```

```python
def carName_url():   #外網所有連結
    root_url = 'https://www.mobile01.com/forumlist.php?f=21'
    res = r.get(root_url,headers=HEADER)
    soup = BeautifulSoup(res.text,'lxml')
    name = soup.select('.subject-text > a')
    car_url = []
    for idx in name:
        if idx.text.upper() in carName_list:
#             print(idx.text)
#             print(host+idx['href'])
            car_url.append(host+idx['href'])
#     print(len(car_url))
#     print(car_url)
    return car_url
```

```python
page_list = []        #從外網爬到總共頁數後  就可以用range(pages) 將內頁網址擷取  並使用re.match篩選
for url in carName_url():
    print(url)
    res = r.get(url,headers=HEADER)
    print("status_code",res.status_code)
    soup = BeautifulSoup(res.text,'lxml')
    page = soup.select('.pagination > a ')
    carName = soup.select(".main > h2")[1].text
    idx = len(page)
    print("page",page[(idx-1)].text)
    print(carName)
    time.sleep(int(random.random()*3+1))
    page_list.append(page[(idx-1)].text)
```
